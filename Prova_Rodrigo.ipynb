## Contexto

Uma plataforma online, que atua na venda de produtos, deseja otimizar sua interface para melhorar a experiência do usuário e aumentar as taxas de conversão. Recentemente, a empresa implementou um redesign de sua página web, e os registros dessa mudança estão detalhados nos dados coletados desde então. As variáveis registradas diariamente incluem o número de visitantes únicos, sessões, taxa de rejeição, páginas por sessão, principais origens do tráfego, tipos de dispositivos usados, eventos ocorridos, taxa de conversão e o tempo médio de sessão.

Essa abordagem visa garantir que as decisões tomadas maximizem as taxas de conversões, fundamentais para o sucesso contínuo da plataforma.

```python
df['mes'] = df['Data'].dt.month
```

```python
df.groupby('mes')['Visitantes Únicos']\
.sum()\
.plot(kind='bar', title='Visitantes Únicos')

plt.ylabel('Visitas x mes')
plt.xlabel('mes')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing


df.groupby('Principais Origens do Tráfego')['Visitantes Únicos']\
.sum()\
.plot(kind='bar', title='Principais Origens do Tráfego')

plt.ylabel('Visitas x mes')
plt.xlabel('mes')

plt.show()
```

```python
label_encoder = preprocessing.LabelEncoder()

df['Principais Origens do Tráfego']= label_encoder.fit_transform(df['Principais Origens do Tráfego'])
```

```python
df['Dispositivos']= label_encoder.fit_transform(df['Dispositivos'])
```

```python
df
df.drop("Data", axis = 1).corr("spearman")

# Heatmap
plt.figure(figsize = (7,7))
sns.heatmap(df.drop("Data", axis = 1).corr("spearman"), annot = True, cmap = "YlGnBu")
plt.title("Mapa de Correlação das Variáveis Numéricas\n", fontsize = 15)
plt.show()
```
